```python
def tod_converter(timestamp):
    converter = {'morning': (5, 13),
                 'afternoon': (13, 16),
                 'evening': (16, 23)}
    
    hour = timestamp.hour
    
    for tod in converter:
        if hour in range(*converter[tod]):
            return tod
    return 'night'
```
